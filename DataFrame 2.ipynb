## Filter a DataFrame with morethan one condition and(|)


```python
filt1=df["Senior Management"]

filt2=df["Start Date"]>"1993-07-31"

```

```python
df[filt1 | filt2]
```

```python
filt1=df["First Name"]=="Henry"
filt2=df["Salary"]>34000
filt3=df["Start Date"]>"2022-01-31"
```

```python
df[filt1 & (filt2 | filt3)]
```

## The .isin() Method

```python
import pandas as pd,numpy as np,os
```

```python
os.chdir('C:/Users/Manjunath/Desktop/Raw_Data')
```

```python
df=pd.read_csv("employees.csv",parse_dates=['Start Date'])
df["Gender"]=df["Gender"].astype("category")
df["Senior Management"]=df["Senior Management"].astype("bool")
df.head(3)
```

```python
filt1=df["Team"]=="Marketing"
filt2=df["Team"]=="Legal"
filt3=df["Team"]=="Human Resources"
```

```python
df[filt1| filt2 | filt3]
```

```python
filt=df["Team"].isin(["Marketing","Legal","Human Resources"])
df[filt].head()
```

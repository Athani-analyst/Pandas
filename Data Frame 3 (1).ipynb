## The .apply( ) Method with Columns Values

```python
def add_mill(number):
    return str(number)+" Millions"
```

```python
columns=["Box_Office","Budget","Bond_Actor_Salary"]

for i in columns:
    bond[i]=bond[i].apply(add_mill)
```

```python
bond.head()
```

```python
def add(string):
    return string+"  Awsome"
```

```python
col=["Actor","Director"]

for i in col:
    bond[i]=bond[i].apply(add)

#bond["Actor"].apply(add)
```

```python
bond.head()
```

## The .apply( ) Method with Rows Values

```python
def good_movie(row):
    actor=row[1]
    budget=row[4]
    
    if actor=="Sean Connery":
        return "The best"
    elif actor=="Roger Moore" and budget >40:
        return "Enjoyable"
    else:
        return "I have no clue"

bond.apply(good_movie, axis="columns")
```

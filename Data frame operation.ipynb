## Drop Rows with Null Value

```python
import pandas as pd,numpy as np,os
```

```python
os.chdir('C:/Users/Manjunath/Desktop/Raw_Data')
```

```python
nba=pd.read_csv("nba.csv")
nba.tail()
```

```python
nba.isnull().sum()
```

```python
nba.dropna(how="any") #how : {'any', 'all'}, default 'any' Determine if row or column is removed from DataFrame, 
                      #when we have at least one NA or all NA.
```

```python
nba.dropna(how="all") # * 'all' : If all values are NA, drop that row or column.
```

```python
nba.dropna(subset='College') #column label or sequence of labels, optional
                              Labels along other axis to consider, e.g. if you are dropping rows
                              these would be a list of columns to include.
```

```python
nba.dropna(subset=['College','Salary'])
```

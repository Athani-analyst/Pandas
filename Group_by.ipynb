## The .groupby() Method

## Retrive A Group with the .get_group( ) Method

```python
fortune[fortune["Sector"]=="Media"] # its filter method but we can done it by groupby?
```

```python
sector=fortune.groupby("Sector")
sector.get_group("Media")
```

```python
sector.get_group("Energy")
sector.get_group("Technology")
sector.get_group("Apparel")
```

## Methods on the Groupby object and DataFrame Columns

```python
sector.max() # it will look the left most column in the DataFrame and Sort it Alphabetically 
             # Max() method give in from descending order
```

```python
sector.min() # it will look the left most column in the DataFrame and Sort it Alphabetically 
             # Max() method give in from ascending order
```

```python
sector.sum()
```

```python
sector.mean()
```

```python
sector["Revenue"].sum()
sector["Employees"].sum()
sector["Profits"].max()
sector["Profits"].min()
sector["Employees"].mean()

sector[["Revenue", "Profits"]].sum()
```

## The .agg( ) Method

```python
fortune.head()
```

```python
sector=fortune.groupby("Sector")
```

```python
fortune.groupby("Sector").agg({"Revenue":"sum","Profits":"mean",'Employees':"sum"})
```

```python
sector.agg(["sum",'mean',"size"])
```

```python
fortune.groupby("Sector").agg({"Revenue":["sum","mean"],
                               "Profits":["mean","sum"],
                               'Employees':"sum"})
```
